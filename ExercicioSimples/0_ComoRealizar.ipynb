# Como realizar os exercícios

- Crie um arquivo `.py` para cada exercício (ex.: `SomaSimples.py`) com a sua própria solução.
  
- Execute localmente no terminal com: `python SomaSimples.py` e teste várias entradas.
  
- Busque e implemente soluções alternativas (uma linha, funções, tratamento de erros, leitura em uma linha, etc.).
  
- Compare suas abordagens com a solução do notebook e anote vantagens/desvantagens.
  
- Salve versões diferentes (ex.: `SomaSimples_v1.py`, `SomaSimples_v2.py`) e comente suas decisões no código.
  
- Use controle de versão (`git`) ou pastas para organizar suas tentativas e facilitar o aprendizado.
