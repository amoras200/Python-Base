## manipulação de arquivos .txt:

```python
# write

with open('example.txt', 'w') as file:
    file.write('Hello, World!')
    file.write('\nThis is an example of writing to a file in Python.')
```

```python
# this "w" mode will overwrite the file if it already exists. If you want to append to the file instead, you can use "a" mode:
with open('example.txt', 'a') as file:
    file.write('\nThis line will be appended to the file.')
```

```python
# read
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```
